```python
def catconsep(df):
    cat = list(df.columns[df.dtypes=='object'])
    con = list(df.columns[df.dtypes!='object'])
    return cat , con
```

```python
cat, con =catconsep(df)
```

```python
df.columns
```

```python
def replacer(df):
    #seperate cat and con
    cat, con = catconsep(df)
    #replacer cat - mode , con- mean/median
    
    for i in df.columns:
        if i in cat:
            md = df[i].fillna(md)
        else: 
            mn = df[i].mean()
            df[i] = df[i].fillna(mn)
      #print
    print('missing values replaced in dataframe')
```

# Categorical Features - oneHotEncoding

```python
X_cat = X[cat]
X_cat.head()
```

```python
df['Cylinders'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
X_cat_pre= ohe.fit_transform(X_cat).toarray()
X_cat_pre
```

```python
cat_cols = ohe.get_feature_names_out()
cat_cols
```

```python
X_cat_pre = pd.DataFrame(X_cat_pre, columns = cat_cols)
X_cat_pre.head()
```

# continuous feature - scalling

```python
X_con = X[con]
X_con.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_con_pre = scaler.fit_transform(X_con)
X_con_pre
```

```python
con_cols = scaler.get_feature_names_out()
con_cols
```

```python
X_con_pre = pd.DataFrame(X_con_pre, columns=con_cols)
X_con_pre.head()
```

```python
X_con_pre.mean()
```

```python
X_con_pre.std()
```

# Combining Cat and Con Pre features


```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

# Performing train test split

80% - training (Pass this into model)

20% - Testing (Unseen to the model)

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, y, test_size=0.2)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

# Creating Linear Regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

# Evaluate model performance with model.score

```python
# R2 score in training
model.score(xtrain, ytrain)
```

```python
# R2 score in testing
model.score(xtest, ytest)
```

# Predicting train and test data

```python
ypred_tr = model.predict(xtrain)
ypred_ts = model.predict(xtest)
```
